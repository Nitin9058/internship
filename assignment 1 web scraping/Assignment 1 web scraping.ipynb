# 1) Write a python program to display all the header tags from wikipedia.org and make data frame.


```python
url='https://en.wikipedia.org/wiki/Main_Page'
page = requests.get(url)
soup = BeautifulSoup(page.content)
header=[i.text for i in soup.find_all(class_="mw-headline")]
df=pd.DataFrame(header,columns=["Heading of Wiki"])
df
```

# 2) Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url='https://www.imdb.com/list/ls055386972/?sort=user_rating,desc&st_dt=&mode=detail&page=1'
IMBD = requests.get(url)
soup=BeautifulSoup(IMBD.content)
movie_name=[i.text.split('\n')[2] for i in soup.find_all( class_="lister-item-header")]
year=[i.text.split('\n')[3] for i in soup.find_all( class_="lister-item-header")]
rating=[i.text.strip() for i in soup.find_all(class_="ipl-rating-star small")]
df=pd.DataFrame({'Name':movie_name,'Rating':rating,'Year':year})
df.head()
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame.
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page5 = requests.get(url)
soup5 = BeautifulSoup(page5.content)
team = soup5.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] 
points = [] 
ratings = [] 
new_list = [] 

for i in soup5.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
for i in soup5.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) 
    points.append(new_list[i+1])
for i in soup5.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
    
df=pd.DataFrame({})
df['Team_name']=team_name[:10]
df['Matches']=matches[:10]
df['Points']=points[:10]
df['Ratings']=ratings[:10]
df    
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame
i) Headline

ii) Time

iii) News Link

```python
News_soup=BeautifulSoup(requests.get("https://www.cnbc.com/world/?region=world").content,"html.parser")
Headline=[i.text for i in News_soup.find_all(class_="LatestNews-headline") ] 
Time=[i.text for i in News_soup.find_all("span",class_="LatestNews-wrapper") ] 
news = News_soup.find_all('div', {'class','LatestNews-container'})
lnk = [each.a.get('href') for each in news]
df=pd.DataFrame({'Headlines':Headline,'Time':Time,"Link":lnk})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in and make data frame

i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image UR

```python
soup=BeautifulSoup(requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special').content,'html.parser')
name=[i.text.split(',')[0] for i in soup.find_all('div',class_='restnt-info cursor')] 
location=[i.text for i in soup.find_all('div',class_='restnt-loc ellipsis')]  
image=[i['data-src'] for i in soup.find_all("img", class_="no-img")] 
cuisine=[i.text.split('|')[1] for i in soup.find_all('span',class_="double-line-ellipsis")] 
rating=[i.text for i in soup.find_all('div',class_='restnt-rating rating-4')] 
df=pd.DataFrame({'Restaurant Name':name,'Cuisine':cuisine,'Location':location,"Rating":rating,"Image_url": image})
df
```
